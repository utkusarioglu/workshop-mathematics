```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols("x")
```

```python
eq = sp.sin(x * sp.cos(x)) * 2
lam_eq = sp.lambdify(x, eq, modules=["numpy"])
x_np = np.linspace(-10, 10, 100)
y_np = lam_eq(x_np)
plt.figure(figsize=(10, 5))
plt.plot(x_np, y_np)
plt.xticks(np.arange(-10, 10, 5))
plt.ylabel("sin of sin")
plt.xlabel("free-range, corn-fed, spain-approved x")
plt.show()
```
