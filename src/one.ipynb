```python
p1 = sp.plot(x*x, show=False)
p2 = sp.plot(x, show=False)
p1.extend(p2)
p1.show()
p2.show()
```

```python
frame =  1.1
sp.plot_implicit(sp.Lt(x**2 + y**2, 1), 
  (x, -1 * frame, frame), (y, -1 * frame, frame), show=True)
```

```python
from sympy.plotting import plot3d
plot3d(x*sp.sin(y**2), (x, -5, 5), (y, -1*0.2*sp.pi, 0.2*sp.pi))
```

```python
sp.plot_implicit(sp.Eq(x, 0.4), (x, -1, 1), (y, -1, 1))
```

```python
l1 = sp.geometry.Segment(sp.Point(1, 1), sp.Point(2, 3))
l2 = sp.geometry.Segment(sp.Point(1, 1), sp.Point(2, 1))
l1.intersection(l2)
```

```python
p = sp.plot_implicit(sp.tan(x) -y, show=False)
kwargs = {
  "show": False,
  "line_color": "gray"
}
for c in range(-3, 4, 2):
  p.extend(sp.plot_implicit(sp.Eq(x, c/2*sp.pi), **kwargs))
p.show()
```
